```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from scipy.interpolate import make_interp_spline, BSpline

x = np.array([240, 245, 250, 255, 260, 265, 270, 275, 280, 285, 290, 295])

ino = np.array([0.476, 0.512, 0.499, 0.394, 0.297, 0.225, 0.171, 0.126, 0.082, 0.033, 0.001, -0.017])
ade = np.array([0.318, 0.44, 0.556, 0.654, 0.686, 0.627, 0.489, 0.332, 0.17, 0.045, 0.006, 0.005])
por = np.array([0.397, 0.471, 0.509, 0.492, 0.432, 0.362, 0.276, 0.203, 0.12, 0.047, 0.008, -0.004])

xnew = np.linspace(x.min(), x.max(), 300)

spl_1 = make_interp_spline(x, ino, k=3) 
power_smooth_1 = spl_1(xnew)

spl_2 = make_interp_spline(x, ade, k=3) 
power_smooth_2 = spl_2(xnew)

spl_3 = make_interp_spline(x, por, k=3)
power_smooth_3 = spl_3(xnew)

plt.plot(xnew, power_smooth_1, label = 'inozyna')
plt.plot(xnew, power_smooth_2, label = 'adenozyna')
plt.plot(xnew, power_smooth_3, label = 'mieszanina poreakcyjna')

plt.xlabel("długość fali (nm)")
plt.ylabel("absorbancja")
plt.legend()
plt.show()

```
